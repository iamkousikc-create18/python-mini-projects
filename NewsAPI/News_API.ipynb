```python
import requests
```

```python
query=input("What type of news you are intrested in? ")
api="2f26b9c454cf42939beae4da2a6d7a22"
url=f"https://newsapi.org/v2/everything?q={query}&from=2025-07-12&sortBy=publishedAt&apiKey={api}"
print(url)
r=requests.get(url)
details=r.json()
print(details)
```

```python
art=details["articles"]
for index,article in enumerate(art):
    print(index+1,article["title"],article["url"])
    print("\n\n")
```
