```python
questions=[["Who is SRK? ","WWE Wrestler","Actor","Astronut","Plumber","b"],
           ["What is the capital city of India? ","Delhi","Kolkata","Mumbai","Bangaluru","a"],
          ["Which Planet is known as the Red Planet? ","Earth","Mars","Venus","Jupiter","c"],
          ["What is the largest mammel? ","Elephant","Blue Whale","Giraffe","Shark","b"],
          ["What is the square root of 64? ","6","3","9","8","d"],
          ["Which Ocean is the largest? ","Indian Ocean","Pacific Ocean","Atlantic Ocean","Arctic Ocean","b"]]
prizes=[100,250,500,1000,10000,50000]
i=0

for question in questions:
    print(question[0])
    print(f"a. {question[1]}")
    print(f"b. {question[2]}")
    print(f"c. {question[3]}")
    print(f"d. {question[4]}")

    #Check weather the answer is correct or not
    a=input("Enter Correct options: ")
    if(question[5]==a):
        print("Correct Answer.Hurray!")
    else:
        print(f"Incorrect answer,Correct Answer is {question[5]}")
        break
    print(f"You Won {prizes[i]}")
    i+=1
```
